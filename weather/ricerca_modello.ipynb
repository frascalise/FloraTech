```python
import pandas as pd
import sklearn as sk
from sklearn.preprocessing import MaxAbsScaler
```

```python
week1_df = pd.read_csv("csv_data\data_union_from14to21_032025.csv",sep=",")
week2_df = pd.read_csv("csv_data\data_union_from21to27_032025.csv",sep=',')

dataset = pd.concat([week1_df, week2_df], axis=0)

columns = dataset.columns

target_attribute = 'precipitation_sum_effettivo'

target = dataset[[target_attribute]]

columns = columns.drop([ target_attribute, 'time', 'showers_sum_effettivo', 
                        'wind_direction_10m_dominant_effettivo', 
                        'wind_gusts_10m_max_effettivo', 
                        'daylight_duration_effettivo',
                        'et0_fao_evapotranspiration_effettivo',
                        'location_id',
                        'temperature_2m_max_effettivo', 'temperature_2m_min_effettivo',
                        'wind_speed_10m_max_effettivo', 'rain_sum_effettivo', 'time.1', 'location_id_effettivo', 'weather_code_effettivo'])


dataset = dataset[columns]


dataset['PREV_DIST'] = [i % 7 for i in range(len(dataset))]

mas = MaxAbsScaler()
new_dataset_transf = mas.fit_transform(dataset)
new_dataset_transf = pd.DataFrame(new_dataset_transf, columns=dataset.columns)

dataset = new_dataset_transf

dataset.head(10)
```

```python

from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.pipeline import Pipeline

from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error
```

```python

X = np.array(dataset)
Y = target.values
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=42)
```

```python
modelSVR = svm.SVR()
modelSVR.fit(X_train, y_train)
modelRandForest = RandomForestRegressor(max_depth=2, random_state=0).fit(X_train, y_train)
modelLinReg = LinearRegression().fit(X_train, y_train)
```

```python
print(LinearRegression().get_params())
print(RandomForestRegressor().get_params())
print(svm.SVR().get_params())
```

```python
results = pd.DataFrame()

results['correct'] = pd.DataFrame(Y)
results['predicted'] = modelSVR.predict(np.array(X))
results['predicted_from_api'] = pd.DataFrame(dataset['precipitation_sum'].values)
results['error'] = abs(results['correct'] - results['predicted'])
results['error_from_api'] = abs(results['correct'] - results['predicted_from_api'])

print("Risultati per SVR:")
print(results['error_from_api'].mean(axis=0))
print(results['error'].mean(axis=0))

resultsLinReg = pd.DataFrame()

resultsLinReg['correct'] = pd.DataFrame(Y)
resultsLinReg['predicted'] = modelLinReg.predict(np.array(X))
resultsLinReg['predicted_from_api'] = pd.DataFrame(dataset['precipitation_sum'].values)
resultsLinReg['error'] = abs(resultsLinReg['correct'] - resultsLinReg['predicted'])
resultsLinReg['error_from_api'] = abs(resultsLinReg['correct'] - resultsLinReg['predicted_from_api'])

print("Risultati per Linear REgression:")
print(resultsLinReg['error_from_api'].mean(axis=0))
print(resultsLinReg['error'].mean(axis=0))


resultsForest = pd.DataFrame()

resultsForest['correct'] = pd.DataFrame(Y)
resultsForest['predicted'] = modelRandForest.predict(np.array(X))
resultsForest['predicted_from_api'] = pd.DataFrame(dataset['precipitation_sum'].values)
resultsForest['error'] = abs(resultsForest['correct'] - resultsForest['predicted'])
resultsForest['error_from_api'] = abs(resultsForest['correct'] - resultsForest['predicted_from_api'])

print("Risultati per Random forest REgression:")
print(resultsForest['error_from_api'].mean(axis=0))
print(resultsForest['error'].mean(axis=0))

```

```python
reg = LinearRegression()
parameters = {"fit_intercept": [True, False], 
    "copy_X": [True, False],
    "positive": [True, False],
             }
grid = GridSearchCV(estimator=reg, param_grid = parameters)
grid.fit(X_train, y_train)

grid.best_params_
```

```python
print("Linear Regression best results:")
optreg = LinearRegression(copy_X=True, fit_intercept=False, positive=True)

optreg.fit(X_train, y_train)

print("MAE: ", mean_absolute_error(np.round(y_test), np.round(optreg.predict(X_test))))
print("MSE: ", mean_squared_error(np.round(y_test), np.round(optreg.predict(X_test))))
print("MSLE: ", mean_squared_log_error(np.round(y_test), np.round(optreg.predict(X_test))))
```

```python
reg = RandomForestRegressor()
parameters = {"n_estimators": [25, 50, 100, 125],
    "criterion": ['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
    "max_depth": [2,3,4],
    "max_features": ['sqrt', 'log2'],
    "warm_start": [True,False],
             }
grid = GridSearchCV(estimator=reg, param_grid = parameters)
grid.fit(X_train, y_train)

trainpred = np.array(grid.predict(X_train))
testpred = np.array(grid.predict(X_test))

print(f"Accuracy on train {sum(abs(y_train - trainpred))}")
print(f"Accuracy on test {sum(abs(y_test - testpred))}")

grid.best_params_
```

```python
for i in grid.best_params_.items():
    print(i)
```

```python
print("Random Forrest Regressior best results:")
optreg = RandomForestRegressor(criterion='absolute_error',max_depth=3,max_features="log2",n_estimators=50,warm_start=True)

optreg.fit(X_train, y_train)

print("MAE: ", mean_absolute_error(np.round(y_test), np.round(optreg.predict(X_test))))
print("MSE: ", mean_squared_error(np.round(y_test), np.round(optreg.predict(X_test))))
print("MSLE: ", mean_squared_log_error(np.round(y_test), np.round(optreg.predict(X_test))))
```

```python
reg = svm.SVR()
parameters = {
    "degree": [2, 3, 4],
    "gamma": ['scale', 'auto'],
    "coef0": [0,0.1,0.25],
    "tol": [0.001, 0.01,0.1],
    "C": [1, 10, 5],
    "epsilon": [0.1, 0.2, 0.25, 0.08],
             }
grid = GridSearchCV(estimator=reg, param_grid = parameters)
grid.fit(X_train, y_train)

grid.best_params_
```

```python
print("Random Forrest Regressior best results:")
optreg = svm.SVR(C=10,coef0=0,degree=2,epsilon=0.25,gamma="scale",tol=0.1)

optreg.fit(X_train, y_train)

print("MAE: ", mean_absolute_error(np.round(y_test), np.round(optreg.predict(X_test))))
print("MSE: ", mean_squared_error(np.round(y_test), np.round(optreg.predict(X_test))))
#print("MSLE: ", mean_squared_log_error(np.round(y_test), np.round(optreg.predict(X_test))))
```
